```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

file_path = r"/Users/philippebao/Desktop/ML_FIRAT/oecd_bli_2015.csv"
df_oecd = pd.read_csv(file_path, thousands= ',')
df_oecd
```

```python
file_path1 = r"/Users/philippebao/Desktop/ML_FIRAT/gdp_per_capita.csv"
df_gdp = pd.read_csv(file_path1, thousands=',', delimiter='\t', encoding='latin1', na_values="n/a")
df_gdp
```

```python
# data preparation
def prepare_country_stats(oecd_bli, gdp_per_capita):

    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]

    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")

    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)

    gdp_per_capita.set_index("Country", inplace=True)

    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,

                                  left_index=True, right_index=True)

    full_country_stats.sort_values(by="GDP per capita", inplace=True)

    remove_indices = [0, 1, 6, 8, 33, 34, 35]

    keep_indices = list(set(range(36)) - set(remove_indices))

    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

```python
country_df= prepare_country_stats(df_oecd, df_gdp)
```

```python
country_df
```

```python
 X=np.c_[country_df["GDP per capita"]]
 Y=np.c_[country_df["Life satisfaction"]]
 Y
```

```python
country_df.plot(kind="scatter", x="Life satisfaction", y="GDP per capita")
plt.show()
```

```python
#Load the linear regression model 
model = LinearRegression()

#Step training of the model 
model.fit(X, Y)
```

```python
# Make a prediction for specific country
X_bis = [[950]]
print(model.predict(X_bis))
```
